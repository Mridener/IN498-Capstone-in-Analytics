# Regression Code

The final CSV file was read into a data frame and used to perform linear and logistic regression. A transformed column was then created, and logistic regression was performed again to compare the results.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
import sys

#Ignoring warnings
if not sys.warnoptions:
  import warnings
  warnings.simplefilter("ignore")

#Set the writeFunction boolean
#True = write to console
#False = write to file for assignment turnin
PRINT = False

#open a file handle for assignment results/
if PRINT == False:
  f = open("IN498_Unit3.txt", "w")
###############################################
##
##PURPOSE: Write to console or file based on
##  the writeFunction variable
##      True = write to concole
##      False = write to file for assignment turnin
##
##INPUT: Message to write (message1)
##   Optional: message2
##
##OUTPUT: None
##
###############################################
def writeFunction(message1, *message2):

  #Print to console
  if PRINT:
    print(message1)
    print(message2)
    print()
  #Print to file for assignment
  else:
    f.write(str(message1))
    f.write(str(message2))
    f.write("\n\n")

#Widen the column display
pd.set_option('max_colwidth',500)

#Read data into a DataFrame using these columns
##"Date","Package_Name","Country","Store_Listing_Visitors",
##"Installers","Visitor-to-Installer_conversion_rate",
##"Installers_retained_for_1_day","Installer-to-1_day_retention_rate",
##"Installers_retained_for_7_days","Installer-to-7_days_retention_rate",
##"Installers_retained_for_15_days","Installer-to-15_days_retention_rate",
##"Installers_retained_for_30_days","Installer-to-30_days_retention_rate"
col_names = ["Date",
             "Package_Name",
             "Country",
             "Store_Listing_Visitors",
             "Installers",
             "Visitor-to-Installer_conversion_rate",
             "Installers_retained_for_1_day",
             "Installer-to-1_day_retention_rate",
             "Installers_retained_for_7_days",
             "Installer-to-7_days_retention_rate",
             "Installers_retained_for_15_days",
             "Installer-to-15_days_retention_rate",
             "Installers_retained_for_30_days",
             "Installer-to-30_days_retention_rate"]

data = pd.read_csv("retained_installers_final.csv")

########### EXPLORE THE DATA SET #######################

#Print the top 10 rows
writeFunction("Top 10 Rows")
writeFunction(data.head(10))

#Print the shape
writeFunction("Shape")
writeFunction(data.shape)

#Print the description of the data
writeFunction("Description")
writeFunction(data.describe(include="all"))

############ FIX MISSING DATA #######################
#Replace NaN with 0 for Installers_retained_for_30_days
data["Installers_retained_for_30_days"] = data["Installers_retained_for_30_days"].fillna(0)


########### LINEAR REGRESSION #######################
#Create a linear regression model
linreg = LinearRegression()

#Set feature_cols to Installers column data only
feature_cols  = ["Installers"]

#Set X to feature_cols
X = data[feature_cols]

#Set y to installer retained for 30 days
y = data ["Installers_retained_for_30_days"]

#Print the shape of X
writeFunction("Shape of X")
writeFunction(X.shape)

#Print the description of X
writeFunction("Description of X")
writeFunction(X.describe())

#Print the shape of y
writeFunction("Shape of y")
writeFunction(y.shape)

#Print the description of y
writeFunction("Description of y")
writeFunction(y.describe())

#Fit the linear regression model with X and y
linreg.fit(X, y)

#Get predictions for retained for 30 days with 1 install
writeFunction("Linear Regression Prediction for 1 Install")
writeFunction(linreg.predict(pd.DataFrame({"Installers": [1]})))

#Get predictions for retained for 30 days with 2 installs
writeFunction("Linear Regression Prediction for 2 Installs")
writeFunction(linreg.predict(pd.DataFrame({"Installers": [2]})))

#Get predictions for retained for 30 days with 4 installs
writeFunction("Linear Regression Prediction for 4 Installs")
writeFunction(linreg.predict(pd.DataFrame({"Installers": [4]})))

#Print the intercept
writeFunction("Linear Regression Intercept")
writeFunction(linreg.intercept_)

#Print the coefficient
writeFunction("Linear Regression Coefficient")
writeFunction(linreg.coef_)

########### LOGISTIC REGRESSION #######################
#Get logistic regression model
logreg = LogisticRegression(max_iter=5000)

#Fit X and y to logistic regression model
logreg.fit(X, y)


#Predict classes using X
pred_class = logreg.predict(X)

#Print the predictions using X
writeFunction("Logistic Regression Predictions")
writeFunction(pred_class)

#Get the predicted probabilities of class 1
class_index = np.where(logreg.classes_ == 1)[0][0]
pred_prob = logreg.predict_proba(X)[:, class_index]

#Print the probabilities using X
writeFunction("Logistic Regression Probabilities of Class 1")
writeFunction(pred_prob)

#Predict the probablity of maintaining
# number of users for 30 days (0, 1, 2 users)

#Get probability for retained users for 30 days with 1 install
writeFunction("Logistic Regression Probability for 1 Install")
writeFunction(logreg.predict_proba(pd.DataFrame({"Installers": [1]}))[:, class_index])

#Get probability for retained users for 30 days with 2 installs
writeFunction("Logistic Regression Probability for 2 Installs")
writeFunction(logreg.predict_proba(pd.DataFrame({"Installers": [2]}))[:, class_index])

#Get probability for retained users for 30 days with 4 installs
writeFunction("Logistic Regression Probability for 4 Installs")
writeFunction(logreg.predict_proba(pd.DataFrame({"Installers": [4]}))[:, class_index])


#################################### INSTALL_30 ##########################################
#Add a new column for installers retained for 30 days. Call it Install_30.
#  If greater than 0, put 1, if 0, put 0
data["Install_30"] = np.where(data["Installers_retained_for_30_days"] > 0, 1, 0)

#Print the top 10 rows of the new data set
writeFunction("Top 10 Rows of New Data Set")
writeFunction(data.head(10))

#Print the shape of the new data set
writeFunction("Shape of New Data Set")
writeFunction(data.shape)

#Print the description of the new data set
writeFunction("Description of New Data Set")
writeFunction(data.describe(include="all"))

########### LOGISTIC REGRESSION #######################

#Perform logistic regression using Install_30 column for y
# X = Installers column
assorted_logreg = LogisticRegression(max_iter=5000)

X = data[["Installers"]]
y = data["Install_30"]

#Fit X and y for logistic regression
assorted_logreg.fit(X, y)

#Print the top 10 rows of X
writeFunction("Top 10 Rows of X")
writeFunction(X.head(10))

#Print the shape rows of X
writeFunction("Shape of X")
writeFunction(X.shape)

#Print the description of X
writeFunction("Description of X")
writeFunction(X.describe())

#Print the top 10 rows of y
writeFunction("Top 10 Rows of y")
writeFunction(y.head(10))

#Print the shape of y
writeFunction("Shape of y")
writeFunction(y.shape)

#Print the description of y
writeFunction("Description of y")
writeFunction(y.describe())

#Predict on X and capture the result to assorted_pred_class
assorted_pred_class = assorted_logreg.predict(X)

#Print the predictions using assorted _pred_class (X predictions)
writeFunction("Assorted Logistic Regression Predictions")
writeFunction(assorted_pred_class)

#Get the predicted probabilities of class 1 and save to assorted_pred_prob
assorted_pred_prob = assorted_logreg.predict_proba(X)[:, 1]

#Print the probabilities using assorted_pred_prob
writeFunction("Assorted Logistic Regression Probabilities")
writeFunction(assorted_pred_prob)

#Predict the probability of maintaining
# a user for 30 days (0, 1)

#Get probability for retained users for 30 days with 1 install
writeFunction("Assorted Logistic Regression Probability for 1 Install")
writeFunction(assorted_logreg.predict_proba(pd.DataFrame({"Installers": [1]}))[:, 1])

#Get probability for retained users for 30 days with 2 installs
writeFunction("Assorted Logistic Regression Probability for 2 Installs")
writeFunction(assorted_logreg.predict_proba(pd.DataFrame({"Installers": [2]}))[:, 1])

#Get probability for retained users for 30 days with 4 installs
writeFunction("Assorted Logistic Regression Probability for 4 Installs")
writeFunction(assorted_logreg.predict_proba(pd.DataFrame({"Installers": [4]}))[:, 1])

#Print the intercept
writeFunction("Assorted Logistic Regression Intercept")
writeFunction(assorted_logreg.intercept_)

#Print the coefficient
writeFunction("Assorted Logistic Regression Coefficient")
writeFunction(assorted_logreg.coef_)

#Close the file handle
if PRINT == False:
    f.close()
```

# Regression Analysis Results

**Compare the Probability Results**

The first logistic regression model predicted a 30-day retention probability of about 14.6% for one install, 34.6% for two installs, and 62.4% for four installs. After the Install_30 column was added, the probabilities increased to about 17.2%, 43.0%, and 90.9%. The second model produced higher probabilities because the data was changed into two groups, retained or not retained after 30 days. This made it easier for the model to predict 30-day retention.

**Intercept Values**

The intercept is the starting point for the model before any installs are included. The linear regression intercept was close to zero, meaning very little retention would be expected with zero installs. The logistic regression intercept was negative, meaning the starting probability of 30-day retention was low before any installs were considered.

**Coefficient Values**

The coefficient shows how the prediction changes when the number of installers increases. In the linear regression model, each additional installer increased the predicted number of retained installers by about 0.32. In the logistic regression model, the positive coefficient showed that the probability of 30-day retention increased as the number of installers increased.

**Summary of the Regression Analysis**

The results from both models showed the same overall trend. As the number of installers increased, the predicted number of retained users and the probability of 30-day retention also increased. The transformed logistic regression model made the retention results easier to separate into retained or not retained after 30 days. Overall, both models showed that higher install counts were associated with higher 30-day retention.

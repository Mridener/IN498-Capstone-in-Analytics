#Parsing CSV Files

```python
import csv

#Files to parse
files = [
    "/content/IN498-Capstone-in-Analytics/retained_installers_com.foo.bar_201904_country.csv",
    "/content/IN498-Capstone-in-Analytics/retained_installers_com.foo.bar_201905_country.csv",
    "/content/IN498-Capstone-in-Analytics/retained_installers_com.foo.bar_201906_country.csv",
    "/content/IN498-Capstone-in-Analytics/retained_installers_com.foo.bar_201907_country.csv",
    "/content/IN498-Capstone-in-Analytics/retained_installers_com.foo.bar_201908_country.csv",
    "/content/IN498-Capstone-in-Analytics/retained_installers_com.foo.bar_201909_country.csv",
    "/content/IN498-Capstone-in-Analytics/retained_installers_com.foo.bar_201910_country.csv",
    "/content/IN498-Capstone-in-Analytics/retained_installers_com.foo.bar_201911_country.csv",
    "/content/IN498-Capstone-in-Analytics/retained_installers_com.foo.bar_201912_country.csv",]

#Column names
columns = [
    "Date",
    "Package_Name",
    "Country",
    "Store_Listing_Visitors",
    "Installers",
    "Visitor-to-Installer_conversion_rate",
    "Installers_retained_for_1_day",
    "Installer-to-1_day_retention_rate",
    "Installers_retained_for_7_days",
    "Installer-to-7_days_retention_rate",
    "Installers_retained_for_15_days",
    "Installer-to-15_days_retention_rate",
    "Installers_retained_for_30_days",
    "Installer-to-30_days_retention_rate"]

#Parse each raw file
for file in files:

  try:
    with open(file, "r", encoding="utf-8-sig") as raw_file:
      lines = raw_file.readlines()

  except UnicodeDecodeError:
    with open(file, "r", encoding="utf-16") as raw_file:
      lines = raw_file.readlines()

  parsed_file = file.replace(".csv", "_parsed.csv")

  #Write parsed CSV file
  with open(parsed_file, "w", newline="", encoding="utf-8") as output_file:

    writer = csv.writer(
        output_file,
        delimiter=",",
        quoting=csv.QUOTE_MINIMAL
      )

    writer.writerow(columns)

    #Parse each row
    for line in lines:
      line = line.strip()
      if line != "":

        #Remove quotes within row
        line = line.strip('"')

        #Split row into columns
        row = line.split(",")

        #Remove quotes from field
        row = [value.strip('"') for value in row]

        writer.writerow(row)

  print(f"Created: {parsed_file}")

print("All nine files were parsed and saved as UTF-8 csv files.")
```

# Data Curation

```python
import pandas as pd

#Read merged dataset
df = pd.read_csv("/content/IN498-Capstone-in-Analytics/retained_installers_final.csv")

#Create copy for cleaning
cleaned_df = df.copy()

#Standardize Date format
cleaned_df["Date"] = pd.to_datetime(cleaned_df["Date"],
                                    errors="coerce",
                                    format="mixed")

cleaned_df["Date"] = cleaned_df["Date"].fillna("Unknown")

#Replace missing Country values
cleaned_df["Country"] = cleaned_df["Country"].fillna("Unknown")

# Remove rows with missing 30-day retention data
cleaned_df = cleaned_df.dropna(subset=["Installers_retained_for_30_days",
                                       "Installer-to-30_days_retention_rate"])

# Save cleaned version as a new file
cleaned_df.to_csv("/content/IN498-Capstone-in-Analytics/retained_installers_final_cleaned.csv",
                  index=False)

print("Cleaned file created.")
print(cleaned_df.shape)
```
